2. Построить круговую диаграмму, демонстрирующую химический состав земной коры в процентном соотношении.
3. На основе dataset о винах построить круговую диаграмму, показывающую распределение производства вин в зависимости от страны происхождения - 10 самых популярных стран.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = Path("data/wine.csv")
df = pd.read_csv(data)
```

```python
"""
Построить столбчатую диаграмму, демонстрирующую соотношение диаметров планет Солнечной системы.
"""

planets = pd.DataFrame(
    {
        "name": [
            "Меркурий",
            "Венера",
            "Земля",
            "Марс",
            "Юпитер",
            "Сатурн",
            "Уран",
            "Нептун",
        ],
        "diameter": [4879, 12104, 12742, 6779, 139820, 116460, 50724, 49244],  # км
    }
)

colors = ["gray", "orange", "blue", "red", "brown", "gold", "lightblue", "darkblue"]

plt.figure(figsize=(10, 6))
plt.bar(
    planets["name"],
    planets["diameter"],
    color=colors,
    edgecolor="black",
    linewidth=1.5,
)
plt.title("Диаметры планет Солнечной системы", fontsize=10, fontweight="bold")
plt.xlabel("Планета", fontsize=10)
plt.ylabel("Диаметр (км)", fontsize=10)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", alpha=0.3, linestyle="--")

for i, (name, diameter) in enumerate(zip(planets["name"], planets["diameter"])):
    plt.text(i, diameter + 3000, f"{diameter:,}", ha="center", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
"""
Построить круговую диаграмму, демонстрирующую химический состав земной коры в процентном соотношении.
"""

elements = [
    "Кислород",
    "Кремний",
    "Алюминий",
    "Железо",
    "Кальций",
    "Натрий",
    "Калий",
    "Магний",
    "Прочие",
]
percentages = [46.6, 27.7, 8.1, 5.0, 3.6, 2.8, 2.6, 2.1, 1.5]

colors = [
    "skyblue",
    "orange",
    "lightcoral",
    "gray",
    "lightgreen",
    "gold",
    "violet",
    "pink",
    "lightgray",
]

plt.figure(figsize=(10, 6))
plt.pie(
    percentages,
    labels=elements,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
)
plt.title("Химический состав земной коры", fontsize=10)
plt.legend(elements, loc="upper right", bbox_to_anchor=(1.2, 1))
plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python
"""
На основе dataset о винах построить круговую диаграмму,
показывающую распределение производства вин в зависимости от страны происхождения.
"""

country_counts = df["country"].value_counts()
top_10_countries = country_counts.head(10)

plt.figure(figsize=(10, 6))
plt.pie(
    np.array(top_10_countries),
    labels=list(top_10_countries.index),
    autopct="%1.1f%%",
    startangle=90,
)
plt.title("Топ-10 стран по производству вин", fontsize=10)
plt.axis("equal")
plt.tight_layout()
plt.show()
```
