Многоуровневая классификация текста на основе лингвистических признаков.

Постройте pipeline, который:
- извлекает признаки через NLTK/spaCy (POS, NER, синтаксические пути, энтропия),
- подаёт их в классификатор (например, XGBoost),
- решает задачу многоуровневой классификации (например, жанр -> поджанр -> тема).
